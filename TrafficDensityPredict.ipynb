```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, BatchNormalization
from sklearn.model_selection import train_test_split
from keras import optimizers
from keras.optimizers import SGD, Adam, RMSprop, Adadelta
import matplotlib.pyplot as plt
from keras.callbacks import History
import keras.initializers
from datetime import datetime
```

```python
TrafficDensityPredict = Sequential()
Unit1=70
Unit2=50
Unit3=60
Unit4=30

TrafficDensityPredict.add(Dense(units=Unit1,input_shape=(3,)))
#TrafficDensityPredict.add(BatchNormalization())
TrafficDensityPredict.add(Activation('elu'))

##TrafficDensityPredict.add(Dense(15,activation='sigmoid', input_dim=3))
##TrafficDensityPredict.add(Dense(45, activation='relu'))
#TrafficDensityPredict.add(Dense(85, activation='sigmoid'))
#TrafficDensityPredict.add(Dense(70, activation='tanh'))
##TrafficDensityPredict.add(Dense(20, activation='sigmoid'))
#TrafficDensityPredict.add(Dense(output_dim=1, activation='linear'))
##TrafficDensityPredict.add(Dense(output_dim=1,activation='relu')
##TrafficDensityPredict.compile(loss='mse', optimizer='Adam')
TrafficDensityPredict.add(Dense(units=Unit2))
#TrafficDensityPredict.add(BatchNormalization())
TrafficDensityPredict.add(Activation('elu'))

TrafficDensityPredict.add(Dense(units=Unit3))
TrafficDensityPredict.add(Activation('elu'))

#TrafficDensityPredict.add(Dense(units=Unit4))
#TrafficDensityPredict.add(BatchNormalization())
#TrafficDensityPredict.add(Activation('relu'))

TrafficDensityPredict.add(Dense(1))
#TrafficDensityPredict.add(BatchNormalization())
TrafficDensityPredict.add(Activation('linear'))
optimizer = SGD(lr=0.001)
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
sgd = SGD(lr=0.001)
TrafficDensityPredict.compile(
    loss='mse', # loss function
    optimizer="Adam", # learning rule
    #optimizer=optimizer, # learning rule
    metrics=['accuracy'] )# show accuracy
```

```python
#TrafficData = pd.read_csv('Traffic.csv')

TrafficData = pd.read_csv('Oneday.csv')

#TrafficData = pd.read_csv('TrafficFlow.csv')
TrafficTrainData,TrafficTestData=train_test_split(TrafficData,test_size=0.1)

TrafficTrainDataX = TrafficTrainData[['Minutes','WeekdayNum','MonthVal']]
#TrafficTrainDataX = TrafficTrainData[['Minutes']]
TrafficTrainDataY = TrafficTrainData['TrafficDensity']

TrafficTestDataX = TrafficTestData[['Minutes','WeekdayNum','MonthVal']]
#TrafficTestDataX = TrafficTestData[['Minutes']]
TrafficTestDataY = TrafficTestData['TrafficDensity']

print(TrafficTrainDataX.shape)
#print(TrafficTestDataX)
plt.scatter(TrafficTrainDataX['Minutes'],TrafficTrainDataY)
plt.show()
```

```python
TrafficDensityPredict.fit(TrafficTrainDataX,TrafficTrainDataY,batch_size=20,epochs=500,validation_split=0.2,shuffle=True)
```

```python
score = TrafficDensityPredict.evaluate(TrafficTestDataX, TrafficTestDataY, verbose=0)
print(score)
```

```python
predictions = TrafficDensityPredict.predict(TrafficTestDataX)
plt.scatter(TrafficTestDataX['Minutes'],predictions)
plt.legend([ 'Predicted Traffic'])
plt.show()
plt.scatter(TrafficTestDataX['Minutes'],TrafficTestDataY,edgecolors='r')

plt.legend([ 'Actual Traffic'])
plt.show()

print(predictions.shape[0])
P = predictions.reshape(predictions.shape[0],)
print(TrafficTestDataY.shape)
print(P.shape)
Error = TrafficTestDataY - P
plt.scatter(TrafficTestDataX['Minutes'],abs(Error))
SumError = sum(abs(Error))
AvgError = SumError/Error.shape[0]
print(AvgError)
plt.show()
```
